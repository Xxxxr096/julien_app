```python
import pandas as pd
```

```python
p = pd.read_csv("fusion_complete_avec_colonnes.csv")
f = pd.read_csv("clean_spv_data.csv")
p.isnull().sum()
```

```python
merged_data = f.merge(
    p[['Matricule', 'Sexe']],
    on='Matricule',
    how='left'
)

merged_data.head()
```

```python
merged_data.isnull().sum()
```

```python
merged_data.to_csv("merged_data_spv.csv",index=False)
```

```python
merged_data["Luc léger"] = merged_data["Luc léger"].str.replace(",", ".", regex=False).astype(float)
```

```python
merged_data.loc[merged_data["Luc léger"] == 0, "Niveau Luc léger"] = 0
```

```python
merged_data.to_csv("merged_data_spv.csv",index=False)
```
